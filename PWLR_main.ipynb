```python
# Main function
def PWLR_main(name, data_type, output_type, k1_num=1, k2_num=1, embed_type = 2, entry_type = 0, grid_type = True, feature_type = False):
    data = PWLR_Embedding.PWLR(name, data_type, output_type=OUTPUT_TYPE)
    n_estimator_list = [10,25,50,100,150,200] #RF n_estimators
    max_depth_list = [None]
    repeat_num=10
    k1 = k1_num
    k2 = k2_num
    if embed_type == 1:
        if entry_type == 0:
            print("--------PWLR-H0 Representation--------")
        elif entry_type == 1:
            print("--------PWLR-H1 Representation--------")
        else:
            print("--------PWLR-H0+H1 Representation--------")
    else:
        embed_type = 2
        if entry_type == 0:
            print("--------PWLR-OPT-H0 Representation--------")
        elif entry_type == 1:
            print("--------PWLR-OPT-H1 Representation--------")
        else:
            print("--------PWLR-OPT-H0+H1 Representation--------")
    print(f"{k1} and {k2}")
    PWLR_h0, PWLR_h1 = data.embed_graph_exact_dataset(weight=True, p=1, error=1e-12, bias=0.0, markov_step=k2, wl_step=k1, embed_type=embed_type)
    result0, result1, result2 = PWLR_classification.grid_search_graph_exact(PWLR_h0, PWLR_h1, name, entry_type, repeat_num, n_estimator_list, max_depth_list,data_type=data_type,embed_type=embed_type, grid_check=grid_type, feature_check=feature_type)
    print(" ")
```

```python
DATA_SET = "MUTAG"
DATA_TYPE = 'Discrete'
OUTPUT_TYPE = 'single'
print("<<"+DATA_SET+">>")
#PWLR-H0 (k1=1, k2=12)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=1, k2_num=12, embed_type=1, entry_type=0, grid_type=True, feature_type=False)
#PWLR-H1 (k1=26, k2=27)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=26, k2_num=27, embed_type=1, entry_type=1, grid_type=True, feature_type=False)
#PWLR-H2 (k1=8, k2=11)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=8, k2_num=11, embed_type=1, entry_type=2, grid_type=True, feature_type=False)
#PWLR-OPT-H0 (k1=1, k2=1)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=1, k2_num=1, embed_type=2, entry_type=0, grid_type=True, feature_type=False)
#PWLR-OPT-H1 (k1=16, k2=27)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=16, k2_num=27, embed_type=2, entry_type=1, grid_type=True, feature_type=False)
#PWLR-OPT-H2 (k1=1, k2=1)
PWLR_main(DATA_SET, DATA_TYPE, OUTPUT_TYPE, k1_num=1, k2_num=1, embed_type=2, entry_type=2, grid_type=True, feature_type=False)
```
